```python
def accuracy_test(classifier, X, y):
    y_pred = classifier.predict(X)
    y_error = y - y_pred
    acc = 1 - ((X.shape[0] - np.where(y_error == 0)[0].shape[0]) 
               / X.shape[0])
    return acc
```

```python
import numpy as np
from sklearn.svm import SVC

for C in [0.1, 1.0, 5.0, 10.0, 50.0]:
    svm_linear = SVC(C=C, kernel='linear')
    svm_linear.fit(X_0, y_0)
    accuracy = accuracy_test(svm_linear, X_0, y_0)
    print('cost = {}, #SV = {}, accuracy = {}'.format(
        C, svm_linear.support_vectors_.shape[0], accuracy))
    
    X_0_sv = X_0[svm_linear.support_]
    y_0_sv = y_0[svm_linear.support_]
    ipython_2d_scatter(X_0_sv, y_0_sv, marker_size=10)
    print('\n{}\n'.format('-'*50))
```

gamma : float, optional (default=auto)

Kernel coefficient for rbf, poly and sigmoid. If gamma is auto then 1/n_features will be used instead.
